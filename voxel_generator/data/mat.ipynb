```python
from scipy.io import loadmat
import json

data = loadmat("ti_te_dict.mat")

clean = {
    key: value.tolist()
    for key, value in data.items()
    if not key.startswith("__")
}

print(json.dumps(clean, indent=2))
```

```python
from scipy.io import loadmat
import json

data = loadmat("ti_te_dict.mat")

clean = {
    key: value.tolist()
    for key, value in data.items()
    if not key.startswith("__")
}

with open("ti_te_dict.json", "w") as f:
    json.dump(clean, f, indent=2)
```
